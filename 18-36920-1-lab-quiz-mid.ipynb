```python
table={('A','Clean'):'Right',('A','Dirty'):'Suck',('B','Clean'):'Right',('B','Dirty'):'Suck',('C','Clean'):'A',('C','Dirty'):'Suck'}
percepts=[]
def simple_reflex_agent(percept):
    print(percept)
    percepts.append(percept)
    action = lookup(percept,table)
    return action

def lookup(percept,table):
    for key in table:
        if key == percept:
            return table[key]

import random
Location = random.choice(['A','B','C'])
Condition = random.choice(['Clean','Dirty'])

while True:
    action= simple_reflex_agent((Location,Condition))
    print(action)
    cmd = input()
    if(cmd == 'Stop!'): break
    if action == 'Right':
        Location = 'B'
        Condition = random.choice(['Clean','Dirty'])
    elif action== 'Right':
        Location = 'A'
        Condition = random.choice(['Clean','Dirty'])
    elif action=='A':
        Location='A'
        condition=random.choice(['Clean','Dirty'])
    
    else:
        Condition = 'Clean'
```
