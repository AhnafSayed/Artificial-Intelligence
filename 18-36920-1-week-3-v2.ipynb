# Table driven agent:

This agent will help a vehicle to change lanes from left to right and right to left by braking when there is a obstacle in front of the vehicle.

```python
table={('Right Lane','No Obstacle'):'Accelerate Forward',('Right Lane','Obstacle'):'Brake and Change Lane',('Left Lane','No Obstacle'):'Accelerate Forward',('Left Lane','Obstacle'):'Brake and Change Lane'}
percepts=[]    
def table_driven_agent(percept):
    print(percept)
    percepts.append(percept)
    action = lookup(percept,table)
    return action

def lookup(percept,table):
    for key in table:
        if key == percept:
            return table[key]

import random
Side = random.choice(['Right Lane','Left Lane'])
State = random.choice(['No Obstacle','Obstacle'])

while True:
    action= table_driven_agent((Side,State))
    print(action)
    cmd = input()
    if(cmd == 'Stop'): break
    if action == 'Accelerate Forward':
        Side = 'Left Lane'
        State = random.choice(['No Obstacle','Obstacle'])
    elif action== 'Brake and Change Lane':
        Side = 'Right Lane'
        State = random.choice(['No Obstacle','Obstacle'])
    else:
        State = 'No Obstacle'   
```

# Simple reflex agent:

```python
def simple_reflex_agent(percept):
    print(percept)
    if percept[1]=='Obstacle':
        action = 'Brake and Change Lane'
    elif percept[0]=='Right Lane':
        action = 'Accelerate Forward'
    elif percept[0]=='Left Lane':
        action = 'Accelerate Forward'
    return action

import random
Side = random.choice(['Right Lane','Left Lane'])
State = random.choice(['No Obstacle','Obstacle'])

while True:
    action= simple_reflex_agent((Side,State))
    print(action)
    cmd = input()
    if(cmd == 'Stop'): break
    if action == 'Accelerate Forward':
        Side = 'Left Lane'
        State = random.choice(['No Obstacle','Obstacle'])
    elif action== 'Brake and Change Lane':
        Side = 'Right Lane'
        State = random.choice(['No Obstacle','Obstacle'])
    else:
        State = 'No Obstacle'

    
```

# Model based agent:

```python
portion = {'Right Lane':'Unknown', 'Left Lane':'Unknown'}
state = 'Improper'
action = 'Unknown'

def update_state(percept,portion):
    global state, action
    if action=='Brake and Change Lane':
        portion[percept[0]]='No Obstacle'
    if portion['Right Lane']=='No Obstacle' and portion['Left Lane']=='No Obstacle':
        state = 'Proper'
    else:
        state = 'Improper'
    return state
        

def model_based_reflex_agent(percept):
    global state,action,portion
    portion[percept[0]]=percept[1]
    state = update_state(percept,portion)
    print('Perception: '+str(percept)) 
    print('Portion: '+str(portion))   
    print('State: '+str(state))  
    if state == 'Proper':
        action = 'Do nothing'
    elif percept[1]=='Obstacle':
        action = 'Brake and Change Lane'
    elif percept[0]=='Right Lane':
        action = 'Change Lane to Left'
    elif percept[0]=='Left Lane':
        action = 'Change Lane to Right'
    return action


import random
Side = random.choice(['Right Lane','Left Lane'])
Status= random.choice(['No Obstacle','Obstacle'])

while True:
    action = model_based_reflex_agent((Side,Status))
    print('Action: '+ str(action))
    if action == 'Change Lane to Left':
        Side = 'Left Lane'
        Status = random.choice(['No Obstacle','Obstacle'])
    elif action == 'Change Lane to Right':
        Side = 'Right Lane'
        Status = random.choice(['No Obstacle','Obstacle'])
    elif action == 'Brake and Change Lane':
        Status = 'No Obstacle'
    elif action == 'Do nothing':
        break
        

    
```
